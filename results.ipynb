```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Tuning Rank and Alpha for LoRA
rank = [16, 32, 64, 128]
accs_1 = [0.8112745098039216, 0.8357843137254902, 0.8627450980392157, 0.8602941176470589]
f1s_1 = [0.8679245283018868, 0.8838821490467937, 0.9017543859649124, 0.9001751313485113]
accs_2 = [0.8357843137254902, 0.8504901960784313, 0.8627450980392157, 0.8700980392156863]
f1s_2 = [0.8834782608695653, 0.8931698774080561, 0.9017543859649124, 0.9071803852889668]

plt.plot(rank, accs_1, label="Accuracy (Scale = 1)")
plt.plot(rank, f1s_1, label="F1 (Scale = 1)")
plt.plot(rank, accs_2, label="Accuracy (Scale = 2)")
plt.plot(rank, f1s_2, label="F1 (Scale=2)")
plt.title("Tuning Rank and Alpha for LoRA (Alpha/Rank = Scale)")
plt.xlabel("Rank")
plt.ylabel("Performance")
plt.legend()
plt.grid()
plt.show()
```
